```python
# Save directories of images and associates masks
img_dir = '/content/Dataset/701_StillsRaw_full'
mask_dir = '/content/Dataset/LabeledApproved_full'

# Define image shape
output_height = 256
output_width = 256
```

Define pairs between images and masks

```python
def make_pair(img, mask):
    pairs_path = []
    pairs = []
    for im in img:
      for ms in mask:
        if im.split(".png")[0] == ms.split("_L.png")[0]:
          pairs_path.append((img_dir + '/' + im, mask_dir + '/' + ms))
          pairs.append((im ,((im.split('.png')[0]) +"_L.png")))
           
       
      
      
    
    return pairs_path
```

```python
# Create file paths from file names
file_list = [f for f in os.listdir(img_dir) if os.path.isfile(os.path.join(img_dir, f))]
mask_list = [f for f in os.listdir(mask_dir) if os.path.isfile(os.path.join(mask_dir, f))]

file_list = sorted(file_list)
mask_list = sorted(mask_list)

pair_list = make_pair(file_list, mask_list)
```

```python
len(set(pair_list))
```

## Create training, validation and test set (50%, 25%, 25%)

```python
# Extract a list of index equal to 50% of the total number of the images 
randomlist = random.sample(range(0, len(pair_list)), int(0.5*len(pair_list)))
```

```python
# Starting from the list of extracted indexes, we divide the indexes of the images that will create the test and the validation set
random_test = []
while len(random_test)<(0.5*len(randomlist)):
  elemento = random.choice(randomlist)
  if elemento not in random_test:
    random_test.append(elemento)

random_val = []
for elemento in randomlist:
  if elemento not in random_test:
    random_val.append(elemento)
```

```python
# Insert the images corresponding to the indexes generated in test and validation set
test_data = []
for i in random_test:
  test_data.append(pair_list[i])

val_data = []
for i in random_val:
  val_data.append(pair_list[i])
```

```python
# Add all the remaining elements left out of test and validation set, in the training set (50% of the initial data)
train_data = []
for i in range(len(pair_list)):
  if i not in randomlist:
    train_data.append(pair_list[i])
```

```python
# Create dataframe for each set in which there are two columns relating respectively to the name of the images of the set (with path) and associated masks (with path)
train_df = pd.DataFrame(train_data, columns = ['train_frame','train_mask'])
test_df = pd.DataFrame(test_data, columns = ['test_frame','test_mask'])
val_df = pd.DataFrame(val_data, columns = ['val_frame','val_mask'])

# Transform the dataframes into .csv and save them in the folder DATA
train_df.to_csv('gdrive/MyDrive/ProgettoDeepLearning/DATA/train_df.csv')
test_df.to_csv('gdrive/MyDrive/ProgettoDeepLearning/DATA/test_df.csv')
val_df.to_csv('gdrive/MyDrive/ProgettoDeepLearning/DATA/val_df.csv')
```
